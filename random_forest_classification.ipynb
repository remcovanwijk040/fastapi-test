# Random Forest Classification

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

# Create pipeline for transforming and training

## Feature Scaling

```python
import joblib
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
# StandardScaler() as scaler
# RandomForestClassifier() as classiefier
```

## Training the Random Forest Classification model on the Training set

```python

pipe = Pipeline([('scaler', StandardScaler()), ('rfs', RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0))])
pipe.fit(X_train, y_train).score(X_test, y_test)
# save
joblib.dump(pipe, "classifier:0.1.0.pkl") 
```
